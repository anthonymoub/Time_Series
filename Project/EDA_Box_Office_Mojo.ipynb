```python
import pandas as pd
df = pd.read_csv("../Data/BoxOfficeMojo/box_office_mojo_cleaned.csv")
df.head()
```

```python
df['top10_gross'] = pd.to_numeric(df['top10_gross'].replace('[\$,]', '', regex=True), errors='coerce')
df = df.dropna(subset=['top10_gross'])
df['top10_gross'] = df['top10_gross'].astype('int')
df = df.sort_values(by=['year' , 'week_no'] , ascending= True)
df
```

```python
import matplotlib.pyplot as plt

# Create the line plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(df['weekend'], df['top10_gross'], linestyle='-', color='b')
plt.title('Top 10 Gross Over Time')
plt.xlabel('Weekend')
plt.ylabel('Top 10 Gross')
plt.grid(True)

# Customize the x-axis labels for better readability
plt.xticks(rotation=45)
plt.xticks(range(len(df['weekend'])), df['weekend'])  # Set custom x-axis labels

# Show the plot
plt.tight_layout()
plt.show()
```
