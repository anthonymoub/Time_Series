```python
import plotly
import pandas as pd

df = pd.read_csv("/Users/anthonymoubarak/Desktop/Time_Series/lab3/covid-hospitalizations.csv")

df.head()
```

```python
df = df[df['entity'] == 'United States']
```

```python
import plotly.express as px

# Assuming 'df' is your DataFrame
fig = px.line(df, x='date', y='value', line_shape='linear', line_dash_sequence=['solid'], color_discrete_sequence=['blue'])

# Customize the figure as needed
fig.update_layout(
    title='Covid Hospitalizations across time (US)',
    xaxis_title='Date',
    yaxis_title='Value',
    title_x=0.5,  # Set the title's x position to center it
)

# Show the plot
fig.show()
```
